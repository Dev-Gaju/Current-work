```python
filenames = os.listdir("train/")
categories = []
for filename in filenames:
    category = filename.split('.')[0]
    if category == 'dog':
        categories.append(1)
    else:
        categories.append(0)
df = pd.DataFrame({
    'filename': filenames,
    'category': categories
})
```

```python
df
```

```python
files =[]
categories =[]
for file_name, cat in zip(df['filename'],df['category']):
    if cat == 0:
        files.append(file_name)
        categories.append(cat)
df1=pd.DataFrame({
    'FileName': files,
    'Categories' : categories
})
df1
```

```python
def makeCsv_Todrive(dest, f, data):
  os.chdir(dest)
  csv_filename =f
  data.to_csv(dest + csv_filename, index=False)
```

```python
label = pd.read_csv("/content/drive/MyDrive/Kaggle/DogBreed/labels.csv")
```

```python
labels= label['breed'].unique()
train =labels[:105]
test =labels[105:]
```

```python
data_list =[]
for index, row in label.iterrows():
  if row[1] in train:
    data_list.append({'breed': row[1], 'id': row[0]})
train_data = pd.DataFrame(data_list)
```

```python
#check the values with mentioned instruction
len(train_data)
```

```python
data_list =[]
for index, row in label.iterrows():
  if row[1] in test:
    data_list.append({'breed': row[1], 'id': row[0]})
test_data = pd.DataFrame(data_list)
```

```python
#check the values with mentioned instruction
len(test_data)
```

```python
def CopyFile(src, dest, data):
  source= src
  os.chdir(src)
  destination = dest
  os.chdir(dest)

  for image_filename in data['id']:
      image_source = os.path.join(source, image_filename + '.jpg')
      image_destination = os.path.join(destination, image_filename + '.jpg')
      shutil.copy(image_source, image_destination)
```

```python
#copy train data
CopyFile("/content/drive/MyDrive/Kaggle/DogBreed/train", "/content/drive/MyDrive/Dataset/DogsVsCat/train/dog", train_data)
```

```python
#used for test data
CopyFile("/content/drive/MyDrive/Kaggle/DogBreed/train", "/content/drive/MyDrive/Dataset/DogsVsCat/test/dog", test_data)
```
